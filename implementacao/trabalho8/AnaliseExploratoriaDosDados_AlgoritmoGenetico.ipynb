## Descrição dos dados

Temos cinco datasets com os nomes apresentados abaixo e com sua descrição:

- `Fat_Supply_Quantity_Data.csv`: Este conjunto de dados inclui a porcentagem de ingestão de gordura de diferentes tipos de alimentos em países ao redor do mundo. - `Protein_Supply_Quantity_Data.csv`: Este conjunto de dados inclui a porcentagem de ingestão de proteínas de diferentes tipos de alimentos em países ao redor do mundo. ### Coluna Undernourished

Temos valores do tipo `<2.5`, que significa que ele está no intervalo de 0 até 2.5, entao acaba que o pandas entende que é uma string e acaba declarando-o como object inves de float. So converter `<2.5` para o valor médio `1.25` e passar para float.*texto em itálico*

```python
#fazendo os reparos como mencionado anteriormente
data_fat['Undernourished'] = data_fat['Undernourished'].str.replace('<2.5', '1.25')
data_fat['Undernourished'] = data_fat['Undernourished'].astype(float)

data_comida_kg['Undernourished'] = data_comida_kg['Undernourished'].str.replace('<2.5', '1.25')
data_comida_kg['Undernourished'] = data_comida_kg['Undernourished'].astype(float)

data_comida_kcal['Undernourished'] = data_comida_kcal['Undernourished'].str.replace('<2.5', '1.25')
data_comida_kcal['Undernourished'] = data_comida_kcal['Undernourished'].astype(float)

data_protein['Undernourished'] = data_protein['Undernourished'].str.replace('<2.5', '1.25')
data_protein['Undernourished'] = data_protein['Undernourished'].astype(float)
```

## 3. Fazendo fusão das tabelas

```python
colunas_em_comum = [
    'Country',
    'Obesity',
    'Confirmed',
    'Deaths',
    'Recovered',
    'Active',
    'Population',
    'Undernourished',
    'Unit (all except Population)',
]

common = data_fat[colunas_em_comum[:-1]]

# remove as colunas em comuns e adiciona (label) ao final do nome de cada coluna
data_fat = data_fat.drop(columns=colunas_em_comum)
data_comida_kg = data_comida_kg.drop(columns=colunas_em_comum)
data_comida_kcal = data_comida_kcal.drop(columns=colunas_em_comum)
data_protein = data_protein.drop(columns=colunas_em_comum)

dfs = [common, data_fat, data_comida_kcal, data_comida_kg, data_protein]
dataset = pd.concat(dfs, axis=1) # concatena as colunas
dataset
```

### Removendo NaNs

Como somente alguns paises possuem valores com NaNs que corresponde a 9% das linhas e são as mesmas dentre todos os datasets, o dataset final não terá essas linhas de NaNs, ou seja, iremos remover do dataset.

```python
dataset.dropna(inplace=True)
```
