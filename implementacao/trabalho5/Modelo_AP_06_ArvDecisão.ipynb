## Técnica utilizada: Árvores de decisão.

### O que são árvores de decisão?

As árvores de decisão podem ser definidas como um grafo acíclico e direcionado, no qual cada nó representa ou um nó de divisão (momento em que é feita uma escolha entre percorrer um ou outro nó), ou um nó folha, conforme o esquema abaixo:



```
from IPython.display import Image, display

display(Image(filename='arvore.png'))
```

**O objetivo de uma árvore de decisão é classificar os dados de maneira que seja possível realizar previsão a partir da base de dados de entrada**

As principais características das árvores de decisão são:

* Utilizam a estratégia dividir para conquistar
* Cada ramo da árvore é  uma conjunção de disjunções
* A precisão é proporcional a quantidade de exemplos de entrada
* Utiliza analises de entropia para avaliar o quão bom é um caminho.



## Principios básicos da Árvore de Decisão

- A **entropia** é o cálculo do ganho de informação baseado em uma medida utilizada na teoria da informação. A entropia caracteriza a (im)pureza dos dados: em um conjunto de dados, é uma medida da falta de homogeneidade dos dados de entrada em relação a sua classificação. Por exemplo, a entropia é máxima (igual a 1) quando o conjunto de dados é heterogêneo.(Mitchell (1997) - Coimbra (2004)).

$$Entropia(S) \equiv -p_{\oplus}log_{2}p_{\oplus} - p_{\ominus}log_{2}p_{\ominus}$$

- O **ganho de informação** mede a redução da entropia causada pela partição dos exemplos de acordo com os valores do atributo.

$$Ganho(S,A) \equiv Entropia(S) - \sum\limits_{v\in val(A)} \frac{\left|S_v\right|}{\left|S\right|}Entropia(S_v)$$



  - O **índice Gini**, desenvolvido por Conrado Gini em 1912, mede o grau de heterogeneidade dos dados. Logo, pode ser utilizado para medir a impureza de um nó, ou em outras palavras, o quanto um nó é heterogeneo (Onoda, 2001)

$$I_G(p) = \sum_{i=1}^{c}{p_i(1 - p_i)} = 1 - \sum_{i=1}^{c}{p_i^2}$$

       Onde: 
       pi é a freqüência relativa de cada classe em cada nó 
       c é o número de classes

- Quando este índice é igual a zero, o nó é puro. Por outro lado, quando ele se aproxima   do   valor   um,   o   nó   é   impuro   (aumenta   o   número   de   classes    uniformemente distribuídas neste nó).


## Descrição do algoritmo escolhido 

## CART (Classification And Regression Trees)

### Características

Tem como objetivo criar suas decisões através de uma divisão estritamente binária com alvo em seus atributos principais e seus percursores, ou seja, ao usar o primeiro nó (raiz da árvore) este só poderá estar ligado a dois nós, assim sucessivamente ao crescimento da árvore, até que encontre uma folha (Wu, 2009).

A cada interação, a árvore é dividade com base na variavel que produzir a maior queda no erro de classificação 

O CART utiliza o Gini Index para medir o índice de impureza do dataset a ser analisado.

A forma com que essa árvore cresce é relacionada à sua quantidade de decisões, gerando divisões binárias até que não existam mais possibilidade. Como resultado, o nó não pode mais ser dividido, este então é transformado em nó terminal(folha).

Dividi o nó em dois nós filhos (as perguntas podem ser apenas aquelas que tem resposta do tipo 'true' ou 'false')

Como nossa base possuí somente dados binários, a utilização deste algoritmo torna-se conveniente

## Motivação para a escolha

### Vantagens

- Fácil de usar e entender.
- Pode lidar com dados categóricos e númericos. 
- Resistente a outliers, por isso requer pouco pré-processamento de dados
- Novos recursos podem ser facilmente adicionados
- Pode ser usado para criar classificadores grandes usando métodos de conjunto.

### Desvantagens

- Propenso a sobreajuste(over-fitting).
- Exigir algum tipo de medição de como eles estão indo.
- É preciso ter cuidado com o ajuste dos parâmetros.
- Pode criar árvores aprendidas tendenciosas se algumas classes dominam.
